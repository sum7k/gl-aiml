```python
X = df.drop(['ZIP Code', 'Personal Loan'], axis=1)
y = df['Personal Loan']
```

### Transform the data i.e. scale / normalize if required

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,robust_scale
scaler = StandardScaler();

colscal=["Experience","Mortgage","Income","CCAvg"]

scaler.fit(X[colscal])
scaled_features = pd.DataFrame(scaler.transform(X[colscal]),columns=colscal)

feature = X.drop(colscal,axis=1)
X = scaled_features.join(feature)
```

### Create the training set and test set in ration of 70:30	

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=.3, random_state = random_state)
print(X_train.shape, X_test.shape)
```

### *Utility function*

```python
from sklearn.metrics import roc_auc_score, roc_curve

def print_stats(y_test, y_predict):
    cm=metrics.confusion_matrix(y_test, y_predict, labels=[1, 0])
    plt.figure()
    sns.heatmap(cm, annot=True, fmt='.2f', xticklabels = ["1", "0"] , yticklabels = ["1", "0"])
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.show()
    
    accuracy = metrics.scorer.accuracy_score(y_test, y_predict)
    precision = metrics.scorer.precision_score(y_test, y_predict)
    recall = metrics.scorer.recall_score(y_test, y_predict)
    f1 = metrics.f1_score(y_test, y_predict)
    roc_auc = roc_auc_score(y_test, y_predict)
    
    print("Accuracy\t:%.2f" % accuracy)
    print("Precision\t:%.2f" % precision)
    print("Recall  \t:%.2f" % recall)
    print("F1 score\t:%.2f" % f1)
    print("roc auc score\t:%.2f" % roc_auc)
    return (accuracy, precision, recall, f1)
```

### Logistic Regression

```python
from sklearn import metrics

from sklearn.linear_model import LogisticRegression

# Fit the model on train
model = LogisticRegression(solver="liblinear", random_state=random_state)
model.fit(X_train, y_train)

y_predict = model.predict(X_test)
y_predict_probabs = model.predict_proba(X_test)

logreg_stats = print_stats(y_test, y_predict)
logreg_roc_curve = roc_curve(y_test, y_predict_probabs[:,1])
```

### Explain Precision, Recall value & F1 Score using the classification report

#### Precision
Precision is the actual number of people who want a personal loan among all people that classifier predicted will want a loan. It is given as TP / (TP + FP).

#### Recall
Among all the people who want a Personal Loan, recall is the number of people identified by the classifier. It is given as TP/(TP + FN).

#### F1 score
F1 score is given as 2 x Precision x Recall/(Precision + Recall) and attains a high value when both precision and recall are high.
